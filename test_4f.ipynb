```python
from utils import *
from matplotlib import pyplot as plt
import numpy as np
```

```python
setup = {
    'N': 8192,
    'L': 8e-3, # 8mm
    'lambda_0': 650e-9, # 650nm, red light
    'f1': 0.045, # 45mm
    'f2': 0.025, # 25mm
    'A': 0.011, # 11mm
    'plot': True
}
N = setup['N']
n = 1024
signal = test_sum(n, a=255, n_cycles=200, sigma=0.1)
field = np.zeros(N)
field[N//2-n//2:N//2+n//2] = signal
```

```python
output = image(field, setup)
```

```python
mse(intensity(output), intensity(field))
```

```python
setup2 = {
    'N': 8192,
    'L': 4e-3, # 8mm
    'lambda_0': 650e-9, # 650nm, red light
    'f1': 0.045, # 45mm
    'f2': 0.025, # 25mm
    'A': 0.011, # 10mm
    'plot': False
}
errors = []

for n in range(128, 8192, 64):
    signal = test_sum(n, a=255, n_cycles=200, sigma=0.1)
    field = np.zeros(N)
    field[N//2-n//2:N//2+n//2] = signal
    output = image(field, setup2)
    errors.append(mse(intensity(output), intensity(field)))
```

```python
plt.plot(errors)
```
